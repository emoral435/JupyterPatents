# In this notebook, we will be analyzing trends within social media patents ever since the start of the 2000's.

```python
# the reason for picking the 2000's is due to my personal choice of wanting to have some data to compare 
# to social media patents before and after the rise of big global social media companies like myspace and facebook
```

```python
# Here, we are going to load the generic g_patent and g_cpc_current DataFrames as well as our pandas, matplotlib etc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set( style = 'white' )
```

```python
# Here is a variable change in the width of the string for the columns. I am placing it here as it may need
# to be changed often, so I will move it right here!
pd.options.display.max_colwidth = 1000
```

```python
df_patent = pd.read_csv("g_patent.tsv", delimiter='\t', dtype={'patent_id': str, 
                                                        'patent_type': str, 
                                                        'patent_title': str,
                                                        'patent_abstract': str,
                                                        'wipo_kind': str,
                                                        'num_claims': int,
                                                        'withdrawn': int,
                                                        'filename': str}, parse_dates=[2])
df_patent.head(20)
```

```python
df_cpc = pd.read_table("g_cpc_current.tsv", delimiter="\t", dtype={"patent_id": int,
                                                               "cpc_sequence": int,
                                                               "cpc_section": str,
                                                               "cpc_subclass": str,
                                                               "cpc_group": str,
                                                               "cpc_type": str,
                                                               "cpc_symbol_position": str})
df_cpc.head(20)
```

```python
# the reason we are going with the technology section as well as the H section (electricity) is due to
# the classification of some digital information being sent via telegraphic communication 
# after running this, it seems as if there is a huge amount of H classes that leads to the bulk of the 
# rows, and makes the overall run time of these few lines take a while
potential_patents = df_cpc.loc[(df_cpc['cpc_section'] == "Y") | (df_cpc['cpc_section'] == "H")]
potential_patents = potential_patents.drop(columns=["cpc_sequence", "cpc_class", "cpc_subclass", "cpc_group", "cpc_symbol_position", "cpc_type"], axis=1)
potential_patents['patent_id'] = potential_patents['patent_id'].astype(str)
potential_patents = pd.merge(df_patent.drop(axis=1, columns=df_patent.columns[5:]), potential_patents, on="patent_id")
```

```python
potential_patents["full_details"] = potential_patents["patent_title"].astype(str) + potential_patents["patent_abstract"].astype(str)
```

```python
potential_patents.drop(columns=["patent_title", "patent_abstract"], axis=1, inplace=True)
```

```python
potential_patents
```

```python
post_2000 = potential_patents[(potential_patents['patent_date'] > "1999-12-31")]
```

```python
all_patents = pd.concat([post_2000.loc[post_2000['full_details'].str.contains('social media', case=False)],
                        post_2000.loc[post_2000['full_details'].str.contains('community', case=False)],
                        post_2000.loc[post_2000['full_details'].str.contains('social network', case=False)],
                        post_2000.loc[post_2000['full_details'].str.contains('profile', case=False)],
                        post_2000.loc[post_2000['full_details'].str.contains('chat', case=False)],
                        post_2000.loc[post_2000['full_details'].str.contains('digital media', case=False)],
                        post_2000.loc[post_2000['full_details'].str.contains('chat', case=False)]
                        ], ignore_index=True)
```

```python
all_patents
```

```python
showcase = all_patents.groupby([all_patents['patent_date'].dt.year.rename('year')]).agg({'count'})
showcase.drop(inplace=True, columns=showcase.columns[1:5])
showcase.plot(kind="line")
plt.legend(["Social Media Related Patents"])
```
